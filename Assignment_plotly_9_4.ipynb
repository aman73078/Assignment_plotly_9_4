```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
# Ans- 
import seaborn as sns
tips = sns.load_dataset("tips")

import plotly.express as px
fig = px.box(data_frame=tips, x = 'day', y = 'total_bill')
fig.show()
```

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
#     the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
#     column with the color parameter.
```

```python
# Ans-
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
import plotly.express as px
fig = px.histogram(data_frame=tips, x = 'sex', y = 'total_bill', pattern_shape='smoker', color='day')
fig.show()
```

```python
# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
#     the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
#       dimensions parameter.
```

```python
# Ans- 
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
import plotly.express as px
fig = px.scatter_matrix(data_frame=iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color='species')
fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
# Ans- We use a displot (also known as a distribution plot) to represent data in histogram form. 

# Example 1-
import plotly.figure_factory as ff
import numpy as np
x1 = np.random.randn(100)
x2 = np.random.randn(100)
x3 = np.random.randn(100)
```

```python
hist_data = [x1,x2,x3]
group_labels = ['Group 1', 'Group 2', 'Group 3']
colors = ['#333F44', '#37AA9C', '#94F3E4']
```

```python
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)
fig.show()
```

```python
import numpy as np
import seaborn as sns
iris = sns.load_dataset("iris")

y1 = np.array(iris['sepal_length'])
y2 = np.array(iris['sepal_width'])
y3 = np.array(iris['petal_length'])
y4 = np.array(iris['petal_width'])
```

```python
hist_data=[y1,y2,y3,y4]
group_labels=['Group 1', 'Group 2', 'Group 3', 'Group 4']
colors = ['red', 'blue','green','yellow']
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors )
fig.show()
```
